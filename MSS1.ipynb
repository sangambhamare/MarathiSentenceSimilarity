```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
# Load your fine-tuned SBERT model
model = SentenceTransformer("sangambhamare/MarathiSentenceSimilarity")

print("Model loaded successfully!")
```

```python
# Cell 2: Similarity Function
def compute_similarity(sent1: str, sent2: str):
    """
    Encode two sentences and return their cosine similarity.
    """
    emb1 = model.encode(sent1, convert_to_tensor=True)
    emb2 = model.encode(sent2, convert_to_tensor=True)
    score = util.pytorch_cos_sim(emb1, emb2).item()
    return {"Sentence Similarity Score": round(score, 4)}
```

```python
# Cell 3: Gradio Interface
import gradio as gr

iface = gr.Interface(
    fn=compute_similarity,
    inputs=[
        gr.Textbox(label="वाक्य १", lines=2, placeholder="ती पुस्तक वाचत आहे."),
        gr.Textbox(label="वाक्य २", lines=2, placeholder="ती पुस्तक वाचन करत आहे.")
    ],
    outputs=gr.Label(num_top_classes=1, label="Similarity Score"),
    title="Marathi Sentence Similarity",
    description="Compute cosine similarity using Sangam Bhamare's fine-tuned SBERT model."
)

iface.launch(share=True)
```
